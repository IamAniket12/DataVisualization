```python

%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
plt.figure()
plt.subplot(1,2,1)
linear_data = np.array([1,2,3,4,5,6,7,8])
plt.plot(linear_data,'-o')
plt.show()
```

```python
expo_data = linear_data**2
plt.subplot(1,2,2)
plt.plot(expo_data,'-o')
```

```python
plt.subplot(1,2,1)
plt.plot(expo_data,'-x')
```

```python
##how can we share axis between two axis

plt.figure()
ax1 = plt.subplot(1,2,1)
plt.plot(linear_data,'-o')
# pass sharey=ax1 to ensure the two subplots share the same y axis
ax2 = plt.subplot(1,2,2,sharey=ax1)
plt.plot(expo_data,'-o')
plt.show()
```

# Histogram

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,sharex = True)
ax  = [ax1,ax2,ax3,ax4]
for n in range(0,len(ax)):
    sample_size = 10**(n+1)
    sample  = np.random.normal(loc=0.0,scale=1.0,size = sample_size)
    ax[n].hist(sample)
    ax[n].set_title('n={}'.format(sample_size))
    
```

# Animation

```python
import matplotlib.animation as animation 

n=120
X = np.random.randn(n)
```

```python
# create the function that will do the plotting, where curr is the current frame
def update(curr):
    # check if animation is at the last frame, and if so, stop the animation a
    if curr==n:
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4,4,0.5)
    plt.hist(X[:curr],bins=bins)
    plt.axis([-4,4,0,30])
    plt.gca().set_title('Sampling the Normal Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate("n= {}".format(curr),[3,27])
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig,update,interval=100)
print(X)
```

# Interactivity

```python
plt.figure()
data = np.random.rand(10)
plt.plot(data)

def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title("Event at pixel {} {} \n and data {} {}".format(event.x,event.y,event.xdata,event.ydata))
   
 # tell mpl_connect we want to pass a 'button_press_event' into onclick when the event is detected
plt.gcf().canvas.mpl_connect('button_press_event',onclick)
```

```python
from random import shuffle
origin = ['China', 'Brazil', 'India', 'USA', 'Canada', 'UK', 'Germany', 'Iraq', 'Chile', 'Mexico']
shuffle(origin)
df = pd.DataFrame({'height':np.random.rand(10),
                   'weight':np.random.rand(10),
                   'origin':origin
                  })
df
```

```python
plt.figure()

plt.scatter(df['height'],df['weight'],picker=5)

plt.gca().set_xlabel('Height')
plt.gca().set_ylabel("weight")
```

```python
def onpick(event):
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title("event came from {}".format(origin))
# tell mpl_connect we want to pass a 'pick_event' into onpick when the event is detected
plt.gcf().canvas.mpl_connect('pick_event',onpick)
```

```python
import matplotlib.gridspec as gridspec
import numpy as np
import pandas as pd
import matplotlib.animation as animation
fig = plt.figure()
gspec = gridspec.GridSpec(2,2)

gspec.update(wspace=0.5, hspace=0.5, left = 0.25, right=1.0, bottom = 0.1, top = 0.8)
x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)+7
x4 = np.random.uniform(14,20, 10000)

normal_plot = plt.subplot(gspec[0,0])
gamaa_plot = plt.subplot(gspec[0,1])
expo_plot = plt.subplot(gspec[1,0])
uniform_plot = plt.subplot(gspec[1,1])

histo = [normal_plot,gamaa_plot,expo_plot,uniform_plot]
X = [x1,x2,x3,x4]
axisNormal = [-9, 3, 0, 1]
axisGamma = [ 0, 15, 0, 1]
axisExponential = [4, 12, 0, 1]
axisUniform = [13, 22, 0 , 0.6]
axeS = [axisNormal, axisGamma, axisExponential, axisUniform]

titles = ['x1 Normal', 'x2 Gamma', 'x3 Exponential', 'x4 Uniform']
colors = ['r', 'k', 'b', 'c']

bins1 = np.arange(-9, 3, 0.4)
bins2 = np.arange(0,10 , 0.4)
bins3 = np.arange(4, 12, 0.4)
bins4 = np.arange(13, 22, 0.4)
bins = [bins1, bins2, bins3, bins4]

def update(curr):
    print("enter")
    if curr == 100:
        a.event_source.stop()
    plt.cla()
    for (histog,bin_item,title,colr,x,axis_all) in zip(histo,bins,titles,colors,X,axeS):
        histog.hist(x[:curr],normed=True,bins=bin_item,color=colr)
        histog.axis(axis_all)
        histog.set_title("n={}".format(curr))
        histog.set_xlabel('Frequency')
        histog.set_ylabel('Value')
a = animation.FuncAnimation(fig,update,interval=100)


   
```
